```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib  
from mpl_toolkits.basemap import Basemap
import seaborn as sns
%matplotlib inline


tra = pd.read_csv('C:\\Users\\Chirantan\\Desktop\\traffic\\Traffic.csv')
tra.columns = ['Taxi_ID', 'DateTime','Longitude','Latitude']             ## Giving name to Columns 
tra            
  
```

```python
def splitDatetime(data) :

   datatime = pd.DatetimeIndex(data.DateTime)  ## Splitting Date-Time column into year, month, day, hour, min, sec
   data['year'] = datatime.year
   data['month'] = datatime.month
   data['day'] = datatime.day
   data['hour'] = datatime.hour
   data['mins'] = datatime.minute
   data['sec'] = datatime.second
   return data

splitDatetime(tra)

tra.head()
```

```python
def Ttime(h,m,s):
    return (h * 3600 + m * 60 + s)/60                     ## Finding out total time using hour, min and sec 
    
tra['Ttime']=Ttime(tra.hour,tra.mins,tra.sec)
tra
```

```python
tra.Longitude.fillna(116.107522)
tra.Latitude.fillna(39.823659)
tra.Ttime.fillna(7.872532e+02)                        ## To fill NA's with appropriate value
tra.Ttime.fillna(7.872532e+02)


i=0
tra['dlo']= tra.Longitude.iloc[i+1]-tra.Longitude
tra['dla']= tra.Latitude.iloc[i+1]-tra.Latitude              ## To find out differece between Latitide and Longitude point 
i=+1                                                            ## considering every next element is the destination


tra.dlo.fillna(4.063179e-01)
tra.dla.fillna(6.824089e-02)

j=0
tra['elo']= tra.dlo.iloc[j]+ tra.Longitude                ## To find out Destination point by adding the difference
tra['ela']= tra.dla.iloc[j] + tra.Latitude
j=+1

tra.elo.fillna(1.161379e+02)
tra.ela.fillna(3.979687e+01)
```

```python
def haversine(lon1, lat1, lon2, lat2,d,c):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2,d,c = map(np.radians, [lon1, lat1, lon2, lat2,d,c])
    # haversine formula                                                         ## Finding out distance using haversine formula
    radius = 6371 
    dlon = d
    dlat = c
    a = np.sin(dlat/2) * np.sin(dlat/2) + np.cos((lat1)) \
        * np.cos((lat2)) * np.sin(dlon/2) * np.sin(dlon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    d =  radius* c

    return d

tra['dist']=haversine(tra.Longitude,tra.Latitude,tra.elo,tra.ela,tra.dlo,tra.dla)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Day')
ax.set_ylabel('Distance')
ax.set_title("Range of Distance Travelled in each Day")      ## Plotting Distance range
ax.scatter(tra['day'],tra['dist']) 
plt.show()
```
